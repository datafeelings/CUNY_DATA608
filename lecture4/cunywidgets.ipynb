```python
#from IPython.html import widgets
#from IPython.display import display
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np
#from plotly.widgets import GraphWidget
%matplotlib inline
import matplotlib.pyplot as plt
import ipywidgets as widgets



```

This example is from [here](https://ueapy.github.io/ipywidgetsinteract-examples.html)

```python


x = np.linspace(0,1,100)
def pltsin(freq, ampl):
    y = ampl*np.sin(2*np.pi*x*freq)
    plt.plot(x, y)
    plt.ylim(-10,10) # fix limits of the vertical axis
    
```

```python
widgets.interact(pltsin, freq=(1,10,0.1), ampl=(1,10,1))
```

And another one with some pet data of mine

```python
hp = pd.read_csv('https://raw.githubusercontent.com/charleyferrari/CUNY_DATA608/master/lecture3/Sample%20Code/hpi.csv')
hp['DATE'] = pd.to_datetime(hp['DATE'])
```

```python
def plotthis(metro, sansa):
    dslice = hp.ix[(hp['Metro'] == metro) & (hp['Seasonality'] == sansa), ['DATE', 'Tier', 'HPI']]
    dslice = dslice.pivot(index = 'DATE', columns = 'Tier', values = 'HPI')
    dslice = dslice.reset_index()
    dslice.plot()
```

```python
widgets.interact(plotthis, metro=list(hp['Metro'].unique()), sansa=list(hp['Seasonality'].unique()))
```
